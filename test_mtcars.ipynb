```python
#데이터와 열을 전달하면 이상값 정보가 출력되는 함수 만들기
def outlier(data, column):
  mean = data[column].mean() #평균
  std = data[column].std() #표준편차
  lowest = mean - (1.5 * std) #하한값(최소 경계값)
  highest = mean + (1.5 * std) #상한값(최대 경계값)
  print('최소 경계값: ', lowest, ' 최대 경계값: ', highest)
  outlier_index = data[column][(data[column] < lowest) | (data[column] > highest)].index #(하한값 < data < 상한값) 만족하는 data index 추출
  return outlier_index
```

```python
print(outlier(X, 'qsec'))
print(X.loc[24, 'qsec'])
X.loc[24, 'qsec'] = 42.4
print(X.loc[24, 'qsec'])
```

```python
print(outlier(X, 'carb'))
print(X.loc[[29, 30], 'carb'])
X.loc[29, 'carb'], X.loc[30, 'carb'] = 5.235, 5.235
print(X.loc[[29, 30], 'carb'])
```

```python
X['am_new'] = X['am'].replace('manual', 0).replace('auto', 1) #수동 인코딩
X = X.drop(columns = 'am') #am 열은 이제 불필요하므로 삭제
print(X.head())
```

```python
condition = X['wt'] < 3.3
X.loc[condition, 'wt_class'] = 0
X.loc[~condition, 'wt_class'] = 1
print(X[['wt', 'wt_class']].head()) #조건을 이용한 wt열에 등급을 부여한 파생변수 wt_class열 생성
X = X.drop(columns = 'wt') #wt열은 불필요하므로 삭제
print(X.head())
```

```python
X['qsec_4'] = X['qsec'] * 4 # qsec 열(1/4 mile 도달시간)에서 4배 한 파생변수 qsec_4 열(1 mile 도달시간)
print(X[['qsec', 'qsec_4']].head())
X = X.drop(columns = 'qsec') #qsec열은 불필요하므로 삭제
print(X.head())
```

```python
from sklearn.model_selection import train_test_split #sklearn 라이브러리에서 데이터 분리하는 함수 가져오기
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.3) #학습 데이터와 테스트 데이터를 7:3 비율로 분할
print(x_train.head())
print(x_test.head())
print(y_train.head())
print(y_test.head())
```

```python
from sklearn.linear_model import LinearRegression #선형회귀분석: sklearn.linear_model 모듈에서 LinearRegression 모델 가져오기
model = LinearRegression()
model.fit(x_train, y_train) #모델에 x_train, y_train 전달하여 선형회귀 모델로 학습
y_train_predicted = model.predict(x_train)
y_test_predicted = model.predict(x_test)
print(model.intercept_) #선형 회귀모델로 도출된y절편
print(model.coef_) #독립변수들의 각 기울기 값
print(model.score(x_train, y_train)) #학습 데이터에 대한 결정계수
print(model.score(x_test, y_test)) #테스트 데이터에 대한 결정계수
```

```python
from sklearn.metrics import r2_score #결정계수(R^2) 함수
from sklearn.metrics import mean_absolute_error #MAE 함수
from sklearn.metrics import mean_squared_error #MSE 함수
import numpy as np # 제곱근 계산을 위한 numpy
print(r2_score(y_train, y_train_predicted)) #결정계수
print(r2_score(y_test, y_test_predicted)) #결정계수
print(mean_squared_error(y_test, y_test_predicted)) #MSE
print(np.sqrt(mean_squared_error(y_test, y_test_predicted))) #RMSE
print(mean_absolute_error(y_test, y_test_predicted)) #MAE
```

```python
from sklearn.ensemble import RandomForestRegressor #앙상블 모듈에서 랜덤 포레스트 회귀 모델 가져오기
model = RandomForestRegressor() #n_estimators(트리의 생성 개수) 기본값 = 100, criterion(기준) 기본값 = MSE
model.fit(x_train, y_train) #랜덤 포레스트 회귀 분석 수행
y_train_predicted = model.predict(x_train) 
y_test_predicted = model.predict(x_test)
print(r2_score(y_train, y_train_predicted)) #결정계수
print(r2_score(y_test, y_test_predicted)) #결정계수
print(mean_squared_error(y_test, y_test_predicted)) #MSE
print(mean_absolute_error(y_test, y_test_predicted)) #MAE
```
