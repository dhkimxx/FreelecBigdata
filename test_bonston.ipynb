```python
import pandas as pd
data = pd.read_csv('boston.csv')
sorted_data = data.sort_values(by = 'MEDV', ascending = True) #data 변수에서 MEDV 칼럼을 기준으로 오름차순 정렬
print(sorted_data['MEDV'].head(10)) #정렬된 데이터에서 MEDV 칼럼만 상위 10개 출력
```

```python
import pandas as pd
data = pd.read_csv('boston.csv')
data_mean = data['RM'].copy() #RM 칼럼 데이터 복사
rm_mean = data_mean.mean() #RM 칼럼 평균
data_mean.fillna(rm_mean, inplace = True) #결측값을 평균값으로 대치
data_del = data['RM'].copy() #RM 칼럼 데이터 복사
data_del.dropna(inplace = True) #결측값 삭제
print(abs(data_mean.std() - data_del.std())) #표준편차의 차를 절댓값으로 출력
```

```python
import pandas as pd
data = pd.read_csv('boston.csv')
zn_mean = data['ZN'].mean()
zn_std = data['ZN'].std()
zn_max = zn_mean + (1.5 * zn_std) #평균에서 표준편차의 1.5배
print(data[data['ZN'] > zn_max]['ZN'].sum()) #평균에서 표준편차의 1.5배보다 큰 ZN 칼럼 데이터의 합
```

```python
import pandas as pd
data = pd.read_csv('boston.csv')
data_IQR = data.drop(columns = ['CHAS', 'RAD']) #CHAS 칼럼과 RAD 칼럼 제외
described_data_IQR = data_IQR.describe().loc['75%'] - data_IQR.describe().loc['25%'] #칼럼별 IQR 추출
print(described_data_IQR) 
#또는 
described_data_IQR_T = data_IQR.describe().iloc[[4, 6]].T #전치행렬을 이용
print(described_data_IQR_T['75%'] - described_data_IQR_T['25%'])
```

```python
import pandas as pd
data = pd.read_csv('boston.csv')
new_data = data['MEDV'].sort_values(ascending = False) #MEDV 칼럼 내림차순 정렬
new_data.iloc[0:28] = new_data.iloc[29]
print(new_data.mean(), new_data.median(), new_data.min(), new_data.max()) #평균, 중위값, 최소값, 최대값 출력
```

```python
import pandas as pd
data = pd.read_csv('boston.csv')
tax_median = data['TAX'].median()
data_new= data[data['TAX'] > tax_median][['CHAS', 'RAD']]
data_new2 = data_new.groupby(['CHAS', 'RAD'])['RAD'].count() #두개의 칼럼을 그룹화 후 카운트 수행
data_new3 = pd.DataFrame(data_new2) #칼럼 이름을 채우기위해 데이터프레임으로 변환
data_new3.columns = ['COUNT'] #칼럼 이름 COUNT로 설정
print(data_new3)
```

```python
import pandas as pd
data = pd.read_csv('boston.csv')
data_asc = data['TAX'].copy()
data_desc = data['TAX'].copy()
data_asc.sort_values(ascending = True, inplace = True) #오름차순, 바로적용
data_desc.sort_values(ascending = False, inplace = True) #내림차순, 바로적용
data_asc.reset_index(drop = True, inplace = True) #인덱스 재설정, 현재 인덱스정보 삭제, 바로적용
data_desc.reset_index(drop = True, inplace = True) #인덱스 재설정, 현재 인덱스정보 삭제, 바로적용
data_concat = pd.concat([data_asc, data_desc], axis = 1) #concat() 이어붙이는 함수
#axis = 0: 행 단위로 이어붙이기, axis = 1 : 칼럼 단위로 이어붙이기 
data_concat['diff'] = abs(data_concat.iloc[:,0] - data_concat.iloc[:,1]) #data_concat 변수의 diff 칼럼값에 저장
print(data_concat['diff'].var()) #분산 구하기
```

```python
#[방법1]
import pandas as pd
data = pd.read_csv('boston.csv')
data2 = round(data['AGE'], 0) #소숫점 첫째자리에서 반올림
data2 = pd.DataFrame(data2) #데이터 프레임으로 변환
data3 = data2.groupby(['AGE'])['AGE'].count() #그룹화하여 그룹별 개수 세기
data3 = pd.DataFrame(data3) #데이터 프레임으로 변환
data3.columns = ['COUNT'] #데이터 프레임의 칼럼 이름 설정
data3.reset_index(inplace = True) #인덱스 초기화
data3.sort_values(by = 'COUNT', ascending = False, inplace = True) #내림차순 정렬
print(data3.iloc[0, 0], data3.iloc[0, 1])

#[방법2]
from scipy.stats import mode #scipy 라이브러리에서 제공하는 mode() 함수를 사용하면 효율적
print(mode(data2))
print(int(mode(data2)[0]), int(mode(data2)[1]))
```
